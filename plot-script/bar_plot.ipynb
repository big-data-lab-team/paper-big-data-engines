```python
import os

import plot

col_names = ["func", "start", "end", "filename", "worker", "thread", "process"]
benchmark_dir = f"../results/benchmarks"
ylim = None
fig_dir = "../latex/figures"
```

# Increment

```python
experiment = "increment"
parameters = ["n_worker", "block_size", "iterations", "delay"]
```

```python
kwargs = {
    "iterations": 5,
    "block_size": 5000,
    "delay": 1.0,
}
xlabel = "# of Nodes"

plot.bar(
    benchmark_dir=benchmark_dir,
    experiment=experiment,
    col_names=col_names,
    parameters=parameters,
    xlabel=xlabel,
    **kwargs,
    title="Increment: Worker",
    ylim=ylim,
)
```

```python
kwargs = {
    "delay": 1.0,
    "block_size": 5000,
    "n_worker": 4,
}
xlabel = "# of Iterations"

plot.bar(
    benchmark_dir=benchmark_dir,
    experiment=experiment,
    col_names=col_names,
    parameters=parameters,
    xlabel=xlabel,
    **kwargs,
    title="Increment: Iteration",
    ylim=ylim,
)
```

```python
kwargs = {
    "iterations": 5,
    "delay": 1.0,
    "n_worker": 4,
}
xlabel = "# of Splits"

plot.bar(
    benchmark_dir=benchmark_dir,
    experiment=experiment,
    col_names=col_names,
    parameters=parameters,
    xlabel=xlabel,
    **kwargs,
    title="Increment: Block Resplit",
    ylim=ylim,
)
```
