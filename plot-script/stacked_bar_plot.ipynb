```python
import os

import plot
from plot import idle_time, wasted_time

col_names = ["func", "start", "end", "filename", "worker", "thread", "process"]
benchmark_dir = f"../results/benchmarks"
fig_dir = "../latex/figures"
```

```python
idle_func = idle_time
```

```python
def func_remap(df):
    df["func"] =  df["func"].replace(
        {
            "dump": "write",
            "load": "read",
        }
    )
    return df
```

# Increment

```python
experiment = "increment"
parameters = ["n_worker", "block_size", "iterations", "delay"]
```

```python
kwargs = {
    "iterations": 5,
    "delay": 1.0,
    "block_size": 5000,
}
xlabel = "# of Nodes"

plot.stacked_bar(
    benchmark_dir=benchmark_dir,
    experiment=experiment,
    col_names=col_names,
    parameters=parameters,
    xlabel=xlabel,
    idle_func=idle_func,
    **kwargs,
    title="Increment: Nodes",
    ylim=180_000,
    func_remap=func_remap,
    save_name=os.path.join(fig_dir, "stacked_increment_worker.jpg"),
)
```

```python
kwargs = {
    "delay": 1.0,
    "n_worker": 4,
    "block_size": 5000,
}
xlabel = "# of Iterations"

plot.stacked_bar(
    benchmark_dir=benchmark_dir,
    experiment=experiment,
    col_names=col_names,
    parameters=parameters,
    xlabel=xlabel,
    idle_func=idle_func,
    **kwargs,
    title="Increment: Iteration",
    ylim=180_000,
    func_remap=func_remap,
    save_name=os.path.join(fig_dir, "stacked_increment_itr.jpg"),
)
```

```python
kwargs = {
    "iterations": 5,
    "n_worker": 4,
    "delay": 1.0,
}
xlabel = "# of Files"

plot.stacked_bar(
    benchmark_dir=benchmark_dir,
    experiment=experiment,
    col_names=col_names,
    parameters=parameters,
    xlabel=xlabel,
    idle_func=idle_func,
    **kwargs,
    title="Increment: Files",
    ylim=180_000,
    func_remap=func_remap,
    save_name=os.path.join(fig_dir, "stacked_increment_block.jpg"),
)
```
